### Meridional Overturning

This notebook is for computing mean meridional overturning circulation. For this, employ the following two methods,

#### Trasient Eluerian Mean (TEM) Framework 

In the TEM framework, we compute the overturning due the mean flow and eddies.

$$ \psi^* =  \overline{\psi} + \psi_e $$

$$ \overline{\psi}(y,z) = - \int_{-H}^{z}\overline{v}(y,z')dz' = - \int_{-H}^{0}\overline{v}(y,z')dz' + \int_{z}^{0}\overline{v}(y,z')dz' $$

$$ \psi_e = \dfrac{1}{N^2}\overline{v'b'} $$

Here, $b = -g\rho/\rho_o$ and $N^2 = \dfrac{\partial \overline{b}}{ \partial z}$. Primes are deviations from the zonal and time averaged quantities. 

#### Output in Rho2 coordinate

In this, we output `vh` field in Rho2 coordinate. We can integrate this diagnostic over `Rho2` to obtain full overturning in density coordinate. We then compute $\psi(y,z)$ from $\psi(y,\rho)$. Eddy component can then be computed as $\psi(y,z) - \overline{\psi}(y,z)$

```python
path = "/work/Hemant.Khatri/MOM6_idealised/PEM_ACC/N_75/Z_star/TS_Wright/Res_025/"

ds_z = xr.open_dataset(path + 'prog_z_150.nc')

ds_z = ds_z.isel(Time=slice(240,300))

ds_rho = xr.open_dataset(path + 'prog_rho2_150.nc') # chunks={'yh':100, 'yq':100})

ds_rho = ds_rho.isel(Time=slice(240,300))

dsgrid = xr.open_dataset(path + "grid.nc", decode_times=False)
dsgrid = dsgrid.drop('Time')

ds_z = xr.merge([ds_z, dsgrid])

ds_rho = xr.merge([ds_rho, dsgrid])
```

```python
print(ds_z)
```

### TEM calculations

```python
grid_z = Grid(ds_z, coords={'X': {'center': 'xh', 'right': 'xq'},
                        'Y': {'center': 'yh', 'right': 'yq'},
                        'Z': {'center': 'z_l', 'outer': 'z_i'} }, periodic=['X']);

g, rho_0 = (9.8, 1035.)
```

```python
# Mean circulation

%time psi_m = (grid_z.cumsum(ds_z['vmo'].mean('Time'), 'Z', boundary='fill'))

psi_m = psi_m.sum('xh')

psi_m = psi_m - psi_m.isel(z_i=35)

%time psi_m = psi_m.load() / rho_0
```

```python
max_r = 10.
depth = ds_z['depth_ocean'].isel(xh=0)

plt.pcolor(ds_z['yq'], ds_z['z_i'], psi_m / 1.e6, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
#CS = plt.contour(ds_z['yh'], ds_z['z_l'], rho, rho_lev, colors='black', linewidths=2)
#plt.clabel(CS, inline=True, fontsize=12)
plt.fill_between(ds_z['yh'], depth, 4000., fc='grey')
plt.ylabel('Depth')
plt.xlabel('Lat')
plt.gca().invert_yaxis()
```

```python
# compute buoyancy and velocity anomalies

b = - g * ds_z['rhopot0'] / rho_0

%time b_m = b.mean(['xh','Time']).load()
b_e = b - b_m

%time v_m = ds_z['v'].mean(['xh','Time']).load()
v_e = ds_z['v'] - ds_z['v'].mean(['xh','Time']) 
```

```python
b_m.plot()
plt.ylim((0.,4000.))
plt.gca().invert_yaxis()
```

```python
# compute buoyancy frequency (-ve sign because xarray sees z positive downwards)
dz = grid_z.diff(ds_z['z_i'], axis='Z', boundary='fill')

N2 = - grid_z.diff(b_m, 'Z', boundary='extend') / grid_z.interp(dz, 'Z', boundary = 'extend')
```

```python
(N2).plot(vmin=-0.0001, vmax=0.0001, cmap='RdBu_r')
plt.ylim((0.,4000.))
plt.gca().invert_yaxis()
```

```python
# compute eddy overturning

psi_e = ((v_e * grid_z.interp(b_e, 'Y', boundary='fill')).mean('Time') 
         / grid_z.interp(grid_z.interp(N2, 'Y', boundary='fill'),'Z', boundary='fill'))

%time psi_e = (grid_z.interp(psi_e,'Z', boundary='fill') * ds_z['dyCv']).sum('xh')
```

```python
psi_e.plot(vmin=-10000000., vmax=10000000.,cmap='RdBu_r')
plt.gca().invert_yaxis()
```

```python
plt.figure(figsize=(18,4))

max_r = 10.
depth = ds_z['depth_ocean'].isel(xh=0)

plt.subplot(1,3,1)
plt.pcolor(ds_z['yq'], ds_z['z_i'], psi_m / 1.e6, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
#CS = plt.contour(ds_z['yh'], ds_z['z_l'], rho, rho_lev, colors='black', linewidths=2)
#plt.clabel(CS, inline=True, fontsize=12)
plt.fill_between(ds_z['yh'], depth, 4000., fc='grey')
plt.ylabel('Depth')
plt.xlabel('Lat')
plt.title('Psi mean')
plt.gca().invert_yaxis()

plt.subplot(1,3,2)
plt.pcolor(ds_z['yq'], ds_z['z_i'], psi_e / 1.e6, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
#CS = plt.contour(ds_z['yh'], ds_z['z_l'], rho, rho_lev, colors='black', linewidths=2)
#plt.clabel(CS, inline=True, fontsize=12)
plt.fill_between(ds_z['yh'], depth, 4000., fc='grey')
plt.ylabel('Depth')
plt.xlabel('Lat')
plt.title('Psi eddy')
plt.gca().invert_yaxis()

psi = (psi_m + psi_e) 
plt.subplot(1,3,3)
plt.pcolor(ds_z['yq'], ds_z['z_i'], psi / 1.e6, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
#CS = plt.contour(ds_z['yh'], ds_z['z_l'], rho, rho_lev, colors='black', linewidths=2)
#plt.clabel(CS, inline=True, fontsize=12)
plt.fill_between(ds_z['yh'], depth, 4000., fc='grey')
plt.ylabel('Depth')
plt.xlabel('Lat')
plt.title('Residual')
plt.gca().invert_yaxis()
```
