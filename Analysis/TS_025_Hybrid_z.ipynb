### Overturning Circulation

```python
ppdir = "/work/Hemant.Khatri/MOM6_idealised/PEM_ACC/N_75/Z_star/TS_Wright/Res_025/"

ds_rho_z = xr.open_dataset(ppdir+"prog_rho2_125.nc")

ppdir2 = "/work/Hemant.Khatri/MOM6_idealised/PEM_ACC/N_75/Hybrid/Res_025/"

ds_rho_hybrid = xr.open_dataset(ppdir2+"progrho2.nc")

dsgrid = xr.open_dataset(ppdir+"grid.nc", decode_times=False)
dsgrid = dsgrid.drop('Time')

ds_rho_z = xr.merge([ds_rho_z, dsgrid])
ds_rho_hybrid = xr.merge([ds_rho_hybrid, dsgrid])
```

```python
grid_r = Grid(ds_rho_z, coords={'X': {'center': 'xh', 'right': 'xq'},
                        'Y': {'center': 'yh', 'right': 'yq'},
                        'Rho': {'center': 'rho2_l', 'outer': 'rho2_i'} }, periodic=['X']);

psi_r_z = grid_r.cumsum(ds_rho_z['vmo'].isel(Time=slice(120,240)).mean('Time'), 'Rho', boundary='fill')
psi_r_z = psi_r_z.sum('xh')

psi_r_z = psi_r_z / 1035.

grid_r = Grid(ds_rho_hybrid, coords={'X': {'center': 'xh', 'right': 'xq'},
                        'Y': {'center': 'yh', 'right': 'yq'},
                        'Rho': {'center': 'rho2_l', 'outer': 'rho2_i'} }, periodic=['X']);

psi_r_hybrid = grid_r.cumsum(ds_rho_hybrid['vmo'].isel(Time=slice(120,240)).mean('Time'), 'Rho', boundary='fill')
psi_r_hybrid = psi_r_hybrid.sum('xh')

psi_r_hybrid = psi_r_hybrid / 1035.
```

```python
plt.figure(figsize=(18,4))

max_r = 0.5e7

plt.subplot(1,3,1)
plt.pcolor(ds_rho_z['yq'], ds_rho_z['rho2_i'], psi_r_z, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
plt.ylabel('Rho2')
plt.title('Z-star')
#if(ct == 2):
plt.xlabel('Lat')
plt.grid()
plt.gca().invert_yaxis()

plt.subplot(1,3,2)
plt.pcolor(ds_rho_hybrid['yq'], ds_rho_hybrid['rho2_i'], psi_r_hybrid, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
plt.ylabel('Rho2')
#if(ct == 2):
plt.xlabel('Lat')
plt.title('Hybrid')
plt.grid()
plt.gca().invert_yaxis()

plt.subplot(1,3,3)
plt.pcolor(ds_rho_z['yq'], ds_rho_z['rho2_i'], psi_r_hybrid - psi_r_z, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
plt.ylabel('Rho2')
plt.title('Difference')
#if(ct == 2):
plt.xlabel('Lat')
plt.grid()
plt.gca().invert_yaxis()
```

```python
z_hybrid = grid_r.cumsum(ds_rho_hybrid['h'].isel(Time=slice(120,240)).mean('Time'), 'Rho', boundary='fill')
z_hybrid = grid_r.interp(z_hybrid, 'Y', boundary='fill').mean('xh')

depth = ds_rho_hybrid['depth_ocean'].isel(xh=0)
plt.pcolor(ds_rho_hybrid['yq'], z_hybrid, psi_r_hybrid, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.fill_between(ds_rho_hybrid['yh'], depth, 4000., fc='grey')
plt.colorbar(extend='both')
plt.ylabel('Depth')
plt.xlabel('Lat')
plt.grid()
plt.gca().invert_yaxis()
```

### Compare Overturning in with and without sponge

```python
ppdir = "/work/Hemant.Khatri/MOM6_idealised/PEM_ACC/N_75/Hybrid/Res_025/"

ds_sponge = xr.open_dataset(ppdir+"progrho2_20.nc")
ds_nosponge = xr.open_dataset(ppdir+"progrho2_nosponge.nc")

dsgrid = xr.open_dataset(ppdir+"grid.nc", decode_times=False)
dsgrid = dsgrid.drop('Time')

ds_sponge = xr.merge([ds_sponge, dsgrid])
ds_nosponge = xr.merge([ds_nosponge, dsgrid])
```

```python
grid_r = Grid(ds_sponge, coords={'X': {'center': 'xh', 'right': 'xq'},
                                 'Y': {'center': 'yh', 'right': 'yq'},
                                 'Rho': {'center': 'rho2_l', 'outer': 'rho2_i'} }, periodic=['X']);

psi_sponge = grid_r.cumsum(ds_sponge['vmo'].isel(Time=slice(120,240)).mean('Time').sum('xh'), 'Rho', boundary='fill')

psi_sponge = psi_sponge - psi_sponge.isel(rho2_i=35)  # To make phi zero at the bottom

psi_sponge = psi_sponge/ 1035.

grid_r = Grid(ds_nosponge, coords={'X': {'center': 'xh', 'right': 'xq'},
                                     'Y': {'center': 'yh', 'right': 'yq'},
                                     'Rho': {'center': 'rho2_l', 'outer': 'rho2_i'} }, periodic=['X']);

psi_nosponge = grid_r.cumsum(ds_nosponge['vmo'].isel(Time=slice(30,60)).mean('Time').sum('xh'), 'Rho', boundary='fill')

psi_nosponge = psi_nosponge - psi_nosponge.isel(rho2_i=35)  # To make phi zero at the bottom

psi_nosponge = psi_nosponge/ 1035.
```

```python
import matplotlib.ticker as ticker

psi_lev = [-3, -2, -1, 2, 4, 5]

max_r = 5.
plt.figure(figsize=(18,4))

plt.subplot(1,2,1)
plt.pcolor(ds_sponge['yq'], ds_sponge['rho2_i'], psi_sponge/1.e6, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
CS = plt.contour(ds_sponge['yq'], ds_sponge['rho2_i'], psi_sponge /1.e6, psi_lev, colors='black', linewidths=2)
fmt = {}
strs = ['-3', '-2', '-1', '2', '4', '5']
for l, s in zip(CS.levels, strs):
    fmt[l] = s
plt.clabel(CS, CS.levels, fmt=fmt, inline=True, fontsize=10)
plt.ylabel('Rho2')
#if(ct == 2):
plt.xlabel('Lat')
plt.grid()
plt.gca().invert_yaxis()
plt.title('With Sponge')

plt.subplot(1,2,2)
plt.pcolor(ds_nosponge['yq'], ds_nosponge['rho2_i'], psi_nosponge/1.e6, vmin = -max_r, vmax = max_r, cmap='RdBu_r')
plt.colorbar(extend='both')
CS = plt.contour(ds_nosponge['yq'], ds_nosponge['rho2_i'], psi_nosponge/1.e6, psi_lev, colors='black', linewidths=2)
fmt = {}
strs = ['-3', '-2', '-1', '2', '4', '5']
for l, s in zip(CS.levels, strs):
    fmt[l] = s
plt.clabel(CS, CS.levels, fmt=fmt, inline=True, fontsize=10)
plt.ylabel('Rho2')
#if(ct == 2):
plt.xlabel('Lat')
plt.grid()
plt.gca().invert_yaxis()
plt.title('Without Sponge')
```
